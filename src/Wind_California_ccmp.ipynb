```python
%matplotlib inline
from netCDF4 import Dataset
import matplotlib.pyplot as plt
import numpy as np
import glob
```

```python
path = glob.glob('../data/ccmp/2010/06/*.nc')
fig_path = '/Users/weiguangwu/research/expansion_fans/figs/201006/'
```

```python
for f in path:
    
    nc = Dataset(f,'r')
    uwnd = nc.variables['uwnd'][:]
    vwnd = nc.variables['vwnd'][:]
    lat = nc.variables['lat'][:]
    lon = nc.variables['lon'][:]
    U = (uwnd**2+vwnd**2)**0.5
    U_mean = np.ma.mean(U,axis= 0)
    ind_lat = (lat>= 22) & (lat<= 47)
    ind_lon = (lon>=220) & (lon<=255)
    U_mean_cal = U_mean[ind_lat,:][:,ind_lon]
    lat_cal = lat[ind_lat]
    lon_cal = lon[ind_lon]
    plt.figure(dpi = 200)
    plt.contourf(lon_cal,lat_cal,U_mean_cal,40)
    plt.colorbar()
    name = f[-21:-13]
    plt.savefig(fig_path + name + '.png',dpi = 300)
    plt.show()
    
```
